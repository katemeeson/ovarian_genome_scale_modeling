```python
import cobra
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
from pandas import DataFrame
import plotly
import plotly.express as px
import plotly.graph_objs as go
import sklearn
import seaborn as sns
import scipy as sp
```

```python
#59M, HEYA8, OV56, CAOV3, COV318, OAW28
labels = ['59M', 'HEYA8', 'OV56', 'CAOV3', 'COV318', 'OAW28']
after_media = [4.68,4.69,5.09,4.68,4.68,4.68]
after_gene_expression = [(32.62-4.68),(15.77-4.69),(20-5.09),(32.67-4.68),(35.13-4.68),(32.12-4.68)]
exp = pd.DataFrame(dict(Experimental_thresholds=[48,16,21.5,46.81,39.3,37]))
width = 0.55       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, after_media, width, label='Media constraints', color = 'silver',edgecolor = 'black')
ax.bar(labels, after_gene_expression, width, bottom=after_media,
       label='Gene expression constraints', color = 'lightskyblue',edgecolor = 'black')
exp['Experimental_thresholds'].plot(kind='line', marker = '+', color='black', ms=10)

ax.set_ylabel('model-predicted doubling time (hours)')
ax.set_title('Model-predicted growth; with experimental thresholds')
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.savefig('growth_plot.png',dpi = 300,bbox_inches='tight')
plt.show()
```

```python
#59M, HEYA8, OV56, CAOV3, COV318, OAW28
labels = ['59M', 'HEYA8', 'OV56', 'CAOV3', 'COV318', 'OAW28']
after_media = [4.68,4.69,5.09,4.68,4.68,4.68]
after_gene_expression = [(32.62-4.68),(34.4-4.69),(30.8-5.09),(32.67-4.68),(35.13-4.68),(32.12-4.68)]
exp = pd.DataFrame(dict(Experimental_growths=[48,16,21.5,46.81,39.3,37]))
width = 0.55       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, after_media, width, label='Media constraints', color = 'silver',edgecolor = 'black')
ax.bar(labels, after_gene_expression, width, bottom=after_media,
       label='Gene expression constraints', color = 'lightskyblue',edgecolor = 'black')
exp['Experimental_growths'].plot(kind='line', marker = '.', color='black', ms=10)

ax.set_ylabel('model-predicted doubling time (hours)')
ax.set_title('Model-predicted growth; no experimental thresholds')
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.savefig('growth_plot_no_fg.png',dpi = 300,bbox_inches='tight')
plt.show()
```
